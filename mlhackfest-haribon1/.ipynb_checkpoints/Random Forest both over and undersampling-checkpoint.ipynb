```python
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train = pd.read_csv('train.csv')

train.head()
```

```python
# Label Encoding the target variable
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

le.fit(train['Scientific Name'])
le_name_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print(le_name_mapping)

train['y'] = le.transform(train['Scientific Name'])
```

```python
train = train.drop('indexId', axis=1)
```

```python
train = train.dropna()

#Setting X and y
y = train['y']
X = train.drop(['Scientific Name', 'y'], axis = 1)
```

## Over-sampling

```python
from imblearn.combine import SMOTEENN
from sklearn.model_selection import train_test_split

RAND = 11

sm = SMOTEENN(sampling_strategy='auto', random_state = RAND)

X_res, y_res = sm.fit_resample(X, y)

# Hold out set of 20%

X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size = 0.2, random_state = RAND)
```

```python
res = pd.DataFrame(y_res, columns=['y'])

res['y'].value_counts()
```

## RFE

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.feature_selection import RFE

rfe = RFE(estimator=RandomForestClassifier(random_state=RAND), n_features_to_select=18, verbose=1)
# rf = RandomForestClassifier(n_estimators=400, min_samples_leaf=0.12, random_state=RAND)

# Fit 'rf' to the training set
rfe.fit(X_train, y_train)

# Predict the test set labels 'y_pred'
y_pred = rfe.predict(X_test)

# Print features used
print('RFE features used:')
print(X.columns[rfe.support_])

# Print features ranked
print('RFE features ranked:')
print(dict(zip(X.columns, rfe.ranking_)))

# Print classification report
print(classification_report(y_test, y_pred))
```

```python
#Random Forest Grid Search
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE
from sklearn.metrics import make_scorer
from sklearn.metrics import f1_score

rf = RandomForestClassifier(random_state=RAND)

params = {'n_estimators': [100, 150, 200, 250, 300], 'max_features':[2, 4, 6]}

scorer = make_scorer(f1_score, average = 'micro')
gs_cv = GridSearchCV(estimator=rf, param_grid=params, scoring=scorer, cv=3)

gs_cv.fit(X_train, y_train)

y_preds_gs = gs_cv.predict(X_test)

# Best Params
print(gs_cv.best_params_)

# Print classification report
print(classification_report(y_test, y_preds_gs))
```

### Feature Engineering, PCA

```python
train = pd.read_csv('train.csv')

train.head()
```

```python
train = train.dropna(axis=0).drop(['indexId'], axis=1)
```

```python
train.columns
```

```python
train['ave_temp'] = (train['Annual Mean Temperature'] - (train['Max Temperature of Warmest Month'] - 
                                                         train['Min Temperature of Coldest Month']))/train['Annual Mean Temperature']
```

```python
train[['ave_temp', 'Annual Mean Temperature']].describe()
```

```python
train['ave_prec'] = (train['Annual Precipitation'] - (train['Precipitation of Wettest Month'] - 
                                                         train['Precipitation of Driest Month']))/train['Annual Precipitation']
```

```python
train[['ave_prec', 'Annual Precipitation']].describe()
```

```python
train['ave_quar_prec'] = (train['Precipitation Seasonality'] - (train['Precipitation of Wettest Quarter'] - 
                                                         train['Precipitation of Driest Quarter']))/train['Precipitation Seasonality']
```

```python
train[['ave_quar_prec', 'Precipitation Seasonality']].describe()
```

```python
train['ave_quar_cold'] = (train['Precipitation Seasonality'] - (train['Precipitation of Coldest Quarter'] - 
                                                         train['Precipitation of Warmest Quarter']))/train['Precipitation Seasonality']
```

```python
train[['ave_quar_cold', 'Precipitation Seasonality']].describe()
```

```python
train['temp_range'] = (train['Temperature Annual Range'] - (train['Mean Temperature of Wettest Quarter'] - 
                                                         train['Mean Temperature of Driest Quarter']))/train['Temperature Annual Range']
```

```python
train[['temp_range', 'Temperature Annual Range']].describe()
```

```python
to_drop = ['Annual Mean Temperature', 'Max Temperature of Warmest Month', 'Min Temperature of Coldest Month',
          'Annual Precipitation', 'Precipitation of Wettest Month', 'Precipitation of Driest Month',
          'Precipitation Seasonality', 'Precipitation of Wettest Quarter', 'Precipitation of Driest Quarter',
          'Precipitation of Coldest Quarter', 'Precipitation of Warmest Quarter',
          'Temperature Annual Range', 'Mean Temperature of Wettest Quarter', 'Mean Temperature of Driest Quarter']

train = train.drop(to_drop, axis=1)
```

```python
train.columns
```

```python
def feature_eng(train):
    train['ave_temp'] = (train['Annual Mean Temperature'] - (train['Max Temperature of Warmest Month'] - 
                                                         train['Min Temperature of Coldest Month']))/train['Annual Mean Temperature']
    train['ave_prec'] = (train['Annual Precipitation'] - (train['Precipitation of Wettest Month'] - 
                                                         train['Precipitation of Driest Month']))/train['Annual Precipitation']
    train['ave_quar_prec'] = (train['Precipitation Seasonality'] - (train['Precipitation of Wettest Quarter'] - 
                                                         train['Precipitation of Driest Quarter']))/train['Precipitation Seasonality']
    train['ave_quar_cold'] = (train['Precipitation Seasonality'] - (train['Precipitation of Coldest Quarter'] - 
                                                         train['Precipitation of Warmest Quarter']))/train['Precipitation Seasonality']
    train['temp_range'] = (train['Temperature Annual Range'] - (train['Mean Temperature of Wettest Quarter'] - 
                                                         train['Mean Temperature of Driest Quarter']))/train['Temperature Annual Range']
    
    to_drop = ['Annual Mean Temperature', 'Max Temperature of Warmest Month', 'Min Temperature of Coldest Month',
          'Annual Precipitation', 'Precipitation of Wettest Month', 'Precipitation of Driest Month',
          'Precipitation Seasonality', 'Precipitation of Wettest Quarter', 'Precipitation of Driest Quarter',
          'Precipitation of Coldest Quarter', 'Precipitation of Warmest Quarter',
          'Temperature Annual Range', 'Mean Temperature of Wettest Quarter', 'Mean Temperature of Driest Quarter']

    a = train.drop(to_drop, axis=1)
    
    return a
```

```python
from imblearn.combine import SMOTEENN
from sklearn.model_selection import train_test_split

RAND = 11

# Label Encoding the target variable
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

le.fit(train['Scientific Name'])
le_name_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print(le_name_mapping)

train['y'] = le.transform(train['Scientific Name'])
train = train.drop('Scientific Name', axis=1)

X = train.drop('y', axis=1)
y = train['y']

#Over and undersampling
sm = SMOTEENN(sampling_strategy='auto', random_state = RAND)

X_res, y_res = sm.fit_resample(X, y)

#Random Forest Model
rf = RandomForestClassifier(random_state=RAND)

#GridSearchCV
params = {'n_estimators': [100, 150, 200, 250, 300], 'max_features':[2, 4, 6]}
scorer = make_scorer(f1_score, average = 'micro')

gs_cv = GridSearchCV(estimator=rf, param_grid=params, scoring=scorer, cv=3)
gs_cv.fit(X_res, y_res)

#Predict on test
test = pd.read_csv('test.csv').set_index('indexId')

#Feature engineering on test
test = feature_eng(test)

#Select best model
model_gs_cv = gs_cv.best_estimator_

test['preds'] = model_gs_cv.predict(test)
test['Predicted'] = le.inverse_transform(test['preds'])
test = test.reset_index()
output_gs_both_fe = pd.DataFrame(test[['indexId', 'Predicted']]).set_index('indexId')

output_gs_both_fe.head()

output_gs_both_fe.to_csv('output_gs_both.csv')
```
